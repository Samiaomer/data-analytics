```python
df[(df['Score'] > 50) & (df['Score'] < 100)]
```

## Modifying a DataFrame

### Remove rows where Score < 0 (invalid data)

```python
df_cleaned = df.drop(df[df['Score'] < 0].index)
df_cleaned
```

### Remove the `City` column

```python
df_cleaned = df_cleaned.drop(columns=['City'])
df_cleaned
```

### Replace missing Age values with the mean Age

```python
df_cleaned['Age'] = df_cleaned['Age'].fillna(df_cleaned['Age'].mean())
df_cleaned
```

### Replace Score values of 0 with the mean Score

```python
df_cleaned['Score'] = df_cleaned['Score'].replace(0, df_cleaned['Score'].mean())
df_cleaned
```

### Remove any remaining rows containing NaN values

```python
df_cleaned = df_cleaned.dropna()
df_cleaned
```

### Set ID as the index label, then remove the ID column

```python
df_cleaned.index = df_cleaned['ID']
df_cleaned = df_cleaned.drop(columns=['ID'])
df_cleaned
```

# Exercise 1.A NumPy: Manipulating Arrays

```python
import numpy as np
```

```python
## Looping Through Arrays
```

```python
arrA = np.array([[11, 12, 13], [14, 15, 16]])
print(arrA)
```

```python
for x in arrA:
    print(x * 2)
```

```python
for x in arrA:
    for y in x:
        print(y * 2)
```

The first loop iterates over each row of arrA and prints the whole row multiplied by 2.
The second loop goes into each individual element in each row and prints each number multiplied by 2.

```python
for x in arrA:
    for y in x:
        loop_result = x * 2  # x*2 multiplies the whole row, not just y
print(loop_result)
# loop_result only shows the last row x2 because the loop keeps overwriting it
```

## Joining and Stacking Arrays

joined_array_1 = np.concatenate((arrA, arrA), axis=0)
print(joined_array_1)

joined_array_2 = np.concatenate((arrA, arrA), axis=1)
print(joined_array_2)

stacked_array_1 = np.stack((joined_array_1, joined_array_1))
print(stacked_array_1)

```python
# This will cause an error - that is expected!
stacked_array_2 = np.stack((joined_array_1, joined_array_2))
# Error: joined_array_1 is shape (4,3) and joined_array_2 is shape (2,6)
# np.stack requires both arrays to have the exact same shape
```

another_joined_array = np.concatenate((joined_array_1, arrA))
print(another_joined_array)

```python
# This will cause an error - that is expected!
another_2 = np.concatenate((joined_array_1, arrA), axis=1)
# Error: joined_array_1 has 4 rows but arrA only has 2 rows
# To join side by side on axis=1, both arrays must have the same number of rows
```

```python
h = np.hstack((arrA, arrA))
print(h)
# hstack joins arrays side by side (left to right)

v = np.vstack((arrA, arrA))
print(v)
# vstack joins arrays on top of each other (up and down)

d = np.dstack((arrA, arrA))
print(d)
# dstack joins arrays along a third depth dimension
```

## Splitting Arrays

```python
splitsky = np.split(joined_array_1, 2)
print(splitsky)
# Result: 2 arrays, each is 2D with 2 rows and 3 columns
```

```python
spliterino = np.split(joined_array_1, 3)
print(spliterino)
# Result: 3 arrays, each is still 2D with 1 row and 3 columns
```

## Searching and Replacing Array Values

```python
result_1 = np.where(arrA % 2 == 0, arrA, 'odd')
print(result_1)
# arrA % 2 == 0 checks if each number is even
# If even: print the number from arrA. If odd: print the word 'odd'
```

## Sorting and Filtering Arrays

```python
arrB = np.array([[7, 6], [13, 5], [7, 3]])
arrB_sorted = np.sort(arrB)
print(arrB_sorted)
```

```python
print(arrB)
z = arrB < 7
arrC = arrB[z]
print(arrC)
```

```python
y = arrB >= 7
arrD = arrB[y]
print(arrD)
```

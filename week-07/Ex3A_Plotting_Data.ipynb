# Exercise 3.A Getting familiar with Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Creating Basic Plots

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, marker='o')
plt.show()
```

```python
plt.plot(x, y, marker='o')
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

## Styling Plots

```python
plt.plot(x, y, linestyle='--', color='red')
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
plt.plot(x, y, linestyle='--', color='red', marker='s', markersize=10, markerfacecolor='green')
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
# Triangle marker, purple color
plt.plot(x, y, linestyle='-', color='purple', marker='^', markersize=10, markerfacecolor='orange')
plt.title('Triangle Markers - Purple & Orange')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
# Diamond marker, blue color
plt.plot(x, y, linestyle='-.', color='blue', marker='D', markersize=10, markerfacecolor='cyan')
plt.title('Diamond Markers - Blue & Cyan')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

# Exercise 3.B More plotting practice

## Advanced Plotting Techniques

```python
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [23, 17, 35, 29]

plt.bar(products, sales)
plt.show()
```

```python
plt.bar(products, sales, color='skyblue', width=0.5)
plt.title('Product Sales')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y)
plt.show()
```

```python
s = np.random.rand(50) * 100
colors = np.random.rand(50)

plt.scatter(x, y, s=s, c=colors, cmap='viridis')
plt.colorbar(label='Color Value')
plt.title('Customized Scatter Plot')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()
```

## Subplots

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

x = np.linspace(0, 10, 100)
y = np.sin(x)
ax1.plot(x, y)

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

x = np.linspace(0, 10, 100)
y = np.sin(x)
ax1.plot(x, y)

data = np.random.randn(1000)
ax2.hist(data)

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Subplot 1: Sine wave
x = np.linspace(0, 10, 100)
y = np.sin(x)
ax1.plot(x, y, color='steelblue')
ax1.set_title('Sine Wave')
ax1.set_xlabel('X values')
ax1.set_ylabel('sin(x)')

# Subplot 2: Histogram
data = np.random.randn(1000)
ax2.hist(data, bins=30, color='salmon', edgecolor='black')
ax2.set_title('Histogram of Random Numbers')
ax2.set_xlabel('Value')
ax2.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

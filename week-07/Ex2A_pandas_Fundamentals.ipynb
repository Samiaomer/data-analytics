```python
import numpy as np
import pandas as pd
```

## Series

```python
# Create four plain Python objects
aa = [1, 2, 3, 4]                                                          # list
ab = (5, 6, 7, 8)                                                          # tuple
ac = {9, 10, 11, 12}                                                       # set
ad = {'first': 13, 'second': 14, 'third': 15, 'fourth': 16}               # dict
```

```python
# Create Series from each of the plain Python objects
ser1 = pd.Series(aa)
ser2 = pd.Series(ab)
ser3 = pd.Series(ac)
ser4 = pd.Series(ad)

print("ser1 (from list):")
print(ser1)
print()
print("ser2 (from tuple):")
print(ser2)
print()
print("ser3 (from set):")
print(ser3)
print()
print("ser4 (from dict):")
print(ser4)
```

```python
# Attempt to create a Series from literal values (not a valid iterable)
# ser5 = pd.Series(17, 18, 19, 20)
# TypeError: Series.__init__() takes from 1 to 2 positional arguments but 5 were given
```

## DataFrames from Series

```python
# Create df1 from ser1
df1 = pd.DataFrame(ser1)
df1
```

```python
# Add two new columns to df1 using insert() with ser2 and ser3
df1.insert(1, 'col2', ser2)
df1.insert(2, 'col3', ser3)
df1
```

```python
# Add a third new column using ser4
df1['col4'] = ser4
df1
# ser4 was created from a dictionary, so it has string labels ('first', 'second', etc.)
# as its index instead of the default integer index (0, 1, 2, 3) used by df1.
# Because the indices don't align, pandas cannot match the values to existing rows,
# so col4 is filled entirely with NaN.
```

```python
# Add a new row at index 4 using loc
df1.loc[4] = [5, 9, 12, 0]
df1
```

## Reading Data into a DataFrame

```python
# Read sakila_example.csv into df2
df2 = pd.read_csv('sakila_example.csv')
df2
```

```python
# Display the top 5 records
df2.head()
```

```python
# Display the last 10 records
df2.tail(10)
```

```python
# Show summary information
df2.info()
```

**Info observations:**
- The output shows the number of rows and columns in the dataset.
- Each column's name, non-null count, and data type (dtype) are listed.
- Columns with fewer non-null values than the total row count contain missing data.
- The dtypes reveal which columns are numeric (int64/float64) vs. text (object), which matters for analysis and aggregation.

```python
# Show column statistics
df2.describe()
```

**Describe observations (sakila):**
- Useful insights: count, mean, min/max, and percentiles for numeric columns give a quick sense of the data's range and distribution. For example, you can spot outliers or skewed distributions from large gaps between the mean and median (50%).
- Less useful: `describe()` only covers numeric columns by default, so any text/categorical columns (like names or categories) are excluded. Summary stats for ID columns (if present) are technically valid but meaningless — averages or standard deviations of IDs carry no real insight.

```python
# Read northwind_example.csv into df3
df3 = pd.read_csv('northwind_example.csv')
df3
```

```python
# Show column statistics for df3
df3.describe()
```

**Describe observations (northwind):**
- Northwind likely has a mix of sales/order data including prices, quantities, and discounts — the describe output surfaces things like average unit price, typical order quantities, and discount ranges, all of which are actionable business metrics.
- Compared to sakila (which is a movie rental schema), northwind's numeric columns tend to be more directly interpretable (e.g., revenue-related figures). The sakila dataset may include more ID columns or rating scales where the statistics are less meaningful.
- If northwind contains date columns stored as strings, `describe()` won't summarize them numerically — highlighting that data type inspection via `info()` should always precede `describe()`.

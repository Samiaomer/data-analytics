## Generating Random Choices from an Array

```python
students = np.array([
    "Samia", "Hashim", "Menal", "Hanan",
    "Nejat", "Selim", "Munir", "Nadia",
    "Fahed", "Amer", "Omer", "Ahmed"
])
```

```python
call_on = np.random.choice(students)
print(call_on)
```

```python
breakout_groups = np.random.choice(students, size=(3, 4))
print(breakout_groups)

# Is this an effective way to assign breakout groups?
# Not really — np.random.choice samples WITH replacement by default,
# meaning the same student could appear in multiple groups simultaneously.
# To fix this, you'd use replace=False, but that only works for a 1-D output.
# For true group assignment without repeats, you'd shuffle the array and reshape it instead.
```

```python
# Probability weights for each of the 12 students:
# Ahmed: 0.00 — zero chance (absent/excused)
# Samia: 0.30 — highest chance, most engaged participant
# Hashim: 0.10
# Selim: 0.10
# Munir: 0.10
# Omer: 0.08
# Fahed: 0.08
# Amer: 0.07
# Nejat: 0.07
# Hanen: 0.05
# Menal: 0.03
# Nadia: 0.02
# Total: 1.00

p_values = [0.00, 0.30, 0.10, 0.10, 0.10, 0.08, 0.08, 0.07, 0.07, 0.05, 0.03, 0.02]

call_on_weighted = np.random.choice(students, size=5, p=p_values)
print(call_on_weighted)
```

## Generating a Normal Distribution

```python
bell_curve = np.random.normal(loc=10, scale=5, size=(4, 5))
print(bell_curve)
```

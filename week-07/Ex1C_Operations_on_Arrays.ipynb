# Exercise 1.C Performing operations on arrays

```python
import numpy as np
```

```python
## Sample Arrays
```

```python
np.random.seed(42)  # optional — makes results reproducible

r_1 = np.random.randint(1, 10, 5)           # 1-D, 5 integers
r_2 = np.random.randint(1, 10, (2, 5))      # 2-D, 2×5
r_3 = np.random.uniform(-10, 10, (2, 4))    # 2-D floats
r_4 = np.random.randint(1, 10, (2, 5))      # 2-D, 2×5 (same shape as r_2)
r_5 = np.random.randint(1, 10, (3, 3))      # 2-D, 3×3

print(r_1, r_2, r_3, r_4, r_5, sep='\n')
```

## Permutations

```python
print("Before shuffle:\n", r_4)
```

```python
np.random.shuffle(r_4)
print("After shuffle:\n", r_4)
```

```python
new_shuffle = np.random.permutation(r_4)
print("new_shuffle:\n", new_shuffle)
print("r_4 (unchanged):\n", r_4)
```

## Mathematical Operations

```python
print("r_1:", r_1, "  shape:", r_1.shape)
print("r_2:\n", r_2, "  shape:", r_2.shape)
result = np.add(r_1, r_2)
print("r_1 + r_2:\n", result, "  shape:", result.shape)
```

```python
print("r_1:", r_1, "  shape:", r_1.shape)
print("r_3:\n", r_3, "  shape:", r_3.shape)
result = np.add(r_1, r_3)
print("r_1 + r_3:\n", result, "  shape:", result.shape)
```

```python
print("r_2:\n", r_2, "  shape:", r_2.shape)
print("r_4:\n", r_4, "  shape:", r_4.shape)
result = np.add(r_2, r_4)
print("r_2 + r_4:\n", result, "  shape:", result.shape)
```

```python
# r_1 shape (5,) vs r_4 shape (2,5) — raises ValueError
# np.add(r_1, r_4)  <-- this would crash because shapes don't match

# r_2 shape (2,5) vs r_4 shape (2,5) — this WORKS, same shape
print(np.add(r_2, r_4))

# adding np.array([1]) to r_4 — broadcasts, adds 1 to everything
print(np.add(np.array([1]), r_4))

# adding np.array([[1]]) to r_4 — same result, still broadcasts
print(np.add(np.array([[1]]), r_4))

# adding np.array([1,1,1,1,1]) to r_4 — adds to each row
print(np.add(np.array([1,1,1,1,1]), r_4))

# adding np.array([[1],[1]]) to r_4 — shape (2,1) broadcasts to (2,5)
print(np.add(np.array([[1],[1]]), r_4))
```

```python
print(np.add(np.array([1]), r_4))       # adds 1 to every element
print(np.add(np.array([[1]]), r_4))     # same — [[1]] broadcasts across all rows/cols
print(np.add(np.array([1,1,1,1,1]), r_4))  # adds element-wise across each row
# np.add(np.array([[1],[1]]), r_4)  ← same shape (2,1)→(2,5) works fine
```

## Rounding

```python
print("r_3:\n", r_3)

print("trunc:\n",  np.trunc(r_3))   # drops decimal, toward 0
print("fix:\n",    np.fix(r_3))     # same as trunc
print("floor:\n",  np.floor(r_3))   # rounds down (toward -∞)
print("ceil:\n",   np.ceil(r_3))    # rounds up (toward +∞)
print("round 2dp:\n", np.round(r_3, 2))  # banker's rounding to 2 places
```

## Cumulative Operations

```python
print("r_4:\n", r_4)

print("Total sum:", np.sum(r_4))
print("Sum along axis 0 (per column):\n", np.sum(r_4, axis=0))
print("Sum along axis 1 (per row):\n",    np.sum(r_4, axis=1))
print("Cumulative sum (flat):\n",         np.cumsum(r_4))
print("Cumulative sum along axis 0:\n",   np.cumsum(r_4, axis=0))
```

## Grocery Problem

```python
# Categories: vegetables, fruits, snacks, beverages
bag1 = np.array([2, 3, 4, 0])
bag2 = np.array([4, 0, 1, 2])
print("bag1:", bag1)
print("bag2:", bag2)
totals = np.sum([bag1, bag2], axis=1)
print(f"The first bag contains {totals[0]} items, "
      f"and the second bag contains {totals[1]} items.")
by_cat = np.add(bag1, bag2)
print(f"There are {by_cat[0]} vegetables, {by_cat[1]} fruits, "
      f"{by_cat[2]} boxes of snacks, and {by_cat[3]} beverages "
      f"in the two grocery bags.")
```

## Teamwork

```python
ali   = np.array([5, 4, 6, 4, 2])
ben   = np.array([0, 5, 5, 3, 4])  # missed Monday → 0
carla = np.array([3, 6, 0, 3, 4])  # missed Wednesday → 0
#Combined array and weekly totals
all_knocks = np.array([ali, ben, carla])
print("all_knocks:\n", all_knocks)

# Per-person total (sum each row)
per_person = np.sum(all_knocks, axis=1)
print(f"Ali: {per_person[0]}, Ben: {per_person[1]}, Carla: {per_person[2]}")
print("Group total:", np.sum(all_knocks)) 

#Rolling estimated votes (every 2 knocks → 1 vote)
# Daily totals for the whole group
daily_knocks = np.sum(all_knocks, axis=0)
print("Daily knocks:", daily_knocks)

# Cumulative total by day, then halve it
rolling_votes = np.cumsum(daily_knocks) * 0.5
print("Rolling estimated votes by day:", rolling_votes) 
```

Shift + Enter

ACME:123-L → supplied by Acme, part #123, size large
DI:12345-M → supplied by DI, part #12345, size medium
ACE:1-12 → supplied by Ace, part #1, size 12

```python
item = "ABC:12345-L"
```

```python
# Function to return all characters before the :

def get_supplier_code(item):
    return item.split(":")[0]

# Test the function
print(get_supplier_code("ABC:12345-L")) 
```

```python
# Function to return all characters between : and -

def get_product_num(item):
    return item.split(":")[1].split("-")[0]

# Test the function
print(get_product_num("ABC:12345-L"))
```

```python
# Function to return all characters after the -

def get_size(item):
    return item.split("-")[1]

# Test the function
print(get_size("ABC:12345-L"))
```

```python
part1 = "ABC:12345-L"
part2 = "XYZ:67890-M"
part3 = "DEF:24680-S"
```

```python
print("For", part1 + ",",
      "the supplier is", get_supplier_code(part1) + ".",
      "Part number", get_product_num(part1),
      "is available in size", get_size(part1) + ".")
```

```python
print("For", part2 + ",",
      "the supplier is", get_supplier_code(part2) + ".",
      "Part number", get_product_num(part2),
      "is available in size", get_size(part2) + ".") 
```

```python
print("For", part3 + ",",
      "the supplier is", get_supplier_code(part3) + ".",
      "Part number", get_product_num(part3),
      "is available in size", get_size(part3) + ".") 
```

```python
# List containing all product codes

parts = ["ABC:12345-L", "XYZ:67890-M", "DEF:24680-S"]

# Loop through the list and display information

for item in parts:
    print("For", item + ",",
          "the supplier is", get_supplier_code(item) + ".",
          "Part number", get_product_num(item),
          "is available in size", get_size(item) + ".")
```

## Lab 3. Employee data

```python
hr_list = [
    ('0123', 'HR', 'Rebecca Yang', '69000'),
    ('0121', 'IT', 'Mark Blick', '67000'),
    ('0068', 'IT', 'Kahna Larsen', '112000'),
    ('0234', 'OPS', 'Jim Smith', '54000')
]
```

```python
# Update Mark Blick to Mark Blick-Hawley

hr_list[1] = ('0121', 'IT', 'Mark Blick-Hawley', '67000') 
```

```python
# Display updated list

print(hr_list) 
```

```python
# Update Jim Smith: OPS → CS and salary → 60000

hr_list[3] = ('0234', 'CS', 'Jim Smith', '60000')
```

```python
# Show final updated list

print(hr_list)
```

```python
# Display updated hr_list in formatted style

for emp in hr_list:
    emp_id = emp[0]
    dept = emp[1]
    name = emp[2]
    salary = int(emp[3])  # convert to integer for formatting

    print(f"{emp_id} | {dept} | {name} | {salary:,}")
```

## Exercise 3 — `favorite_things()`
A function that accepts a name, a favorite movie, and a favorite musician/band, then returns three lines of formatted text.

```python
def favorite_things(name, movie, music):
   
    return (
        f"Name:              {name}\n"
        f"Favorite Movie:    {movie}\n"
        f"Favorite Musician: {music}"
    )


print(favorite_things(
    name  = "Samia Omer",
    movie = "Layl",
    music = "Elysa"
))
```

## Exercise 5 — `favorite_place()`
A function with two parameters — `place` and `why` — that outputs a sentence describing a favorite place to visit and the reason why.

```python
def favorite_place(place, why):
   
    return f"One of my favorite places is {place} because {why}."
```

```python
# --- Test call 1 ---
print(favorite_place(
    place = "Dubai",
    why   = "Dubai is one of my favorite places because the skyline, luxury, and culture make every visit unforgettable"
))
```

```python
# --- Test call 2 ---
print(favorite_place(
    place = "Asmara Eritrea",
    why   = "Asmara is one of my favorite places because I was born and raised there, and no matter where life takes me, it will always feel like home"
))
```

## Exercise 3 — `favorite_things()`
A function that accepts a name, a favorite movie, and a favorite musician/band, then returns three lines of formatted text.

```python
def favorite_things(name, movie, music):
    """
    Returns three lines listing the person's name,
    their favorite movie, and their favorite musician/band.

    Parameters:
        name  (str): Your full name
        movie (str): Your favorite movie
        music (str): Your favorite musician or band

    Returns:
        str: A formatted multi-line string
    """
    return (
        f"Name:              {name}\n"
        f"Favorite Movie:    {movie}\n"
        f"Favorite Musician: {music}"
    )


print(favorite_things(
    name  = "Alex Johnson",
    movie = "Inception",
    music = "Radiohead"
))
```

## Exercise 4 — `why_im_here()`
A function with a single parameter `purpose` that returns a sentence describing why you joined the Year Up United Data Analyst program.

```python
def why_im_here(purpose):
    """
    Returns the value supplied to purpose — a sentence describing
    why the caller joined the Year Up United Data Analyst program.

    Parameters:
        purpose (str): A sentence explaining your motivation

    Returns:
        str: The purpose string as provided
    """
    return purpose


print(why_im_here(
    purpose = "I joined the Year Up United Data Analyst program to build "
              "technical skills that will help me launch a career in data "
              "and create new opportunities for myself and my community."
))
```

## Exercise 5 — `favorite_place()`
A function with two parameters — `place` and `why` — that outputs a sentence describing a favorite place to visit and the reason why.

```python
def favorite_place(place, why):
    """
    Returns a sentence describing a favorite place and the reason why.

    Parameters:
        place (str): Name of the favorite place
        why   (str): Reason it is a favorite

    Returns:
        str: A formatted sentence
    """
    return f"One of my favorite places is {place} because {why}."
```

```python
# --- Test call 1 ---
print(favorite_place(
    place = "Yosemite National Park",
    why   = "the towering granite cliffs and peaceful hiking trails always clear my mind"
))
```

```python
# --- Test call 2 ---
print(favorite_place(
    place = "New Orleans",
    why   = "the vibrant music scene, rich culture, and incredible food are unlike anywhere else"
))
```

## Exercise 6 — Peer Review Reflection

After sharing code with your peer group, consider the following questions:

| Question | Things to think about |
|---|---|
| **Different approaches** | Did peers use `return` vs. `print()` inside the function? String concatenation vs. f-strings? |
| **Most efficient** | Fewer lines with the same readable output generally wins. F-strings tend to be concise and clear. |
| **Most readable / reusable** | Functions that only `return` a value (no `print` inside) are more reusable — the caller decides how to use the output. Descriptive parameter names and a docstring make functions easier for others to understand and reuse. |

> **Key takeaway:** A function that *returns* a value is more flexible than one that *prints* internally,  
> because the returned value can be printed, stored in a variable, passed to another function, or written to a file.

# Tax Functions – Lab Exercise

## Description
This notebook demonstrates how to define and call **reusable Python functions** for payroll tax calculations.

Three helper functions are created, each accepting a gross pay amount and returning the corresponding tax:

| Function | Tax | Rate |
|---|---|---|
| `get_soc_sec_tax()` | Social Security | 6.2% |
| `get_medicare_tax()` | Medicare | 1.45% |
| `get_federal_tax()` | Federal withholding | 18%–23% depending on withholding code |

The federal tax function uses an **if/else** statement to apply the correct rate based on the employee's withholding code:
- Code 0 → 23%
- Code 1 → 21%
- Code 2 → 19.5%
- Code 3 → 18.5%
- Code 4+ → 18%, minus 0.5% for each withholding above 4

The functions are then tested against three sample employees to verify correct output.

```python
# Function 1 – Social Security Tax
def get_soc_sec_tax(gross_pay):
    """Return Social Security tax (6.2%) on the given gross pay."""
    SOC_SEC_RATE = 0.062
    return gross_pay * SOC_SEC_RATE
```

```python
# Function 2 – Medicare Tax
def get_medicare_tax(gross_pay):
    """Return Medicare tax (1.45%) on the given gross pay."""
    MEDICARE_RATE = 0.0145
    return gross_pay * MEDICARE_RATE
```

```python
# Function 3 – Federal Withholding Tax
def get_federal_tax(gross_pay, withholding_code):
    """
    Return federal withholding tax on the given gross pay.

    Withholding code → tax rate:
        0  → 23.0%
        1  → 21.0%
        2  → 19.5%
        3  → 18.5%
        4+ → 18.0%, minus 0.5% for each withholding over 4
               e.g. code 6: 0.18 - (2 × 0.005) = 0.17 (17%)
    """
    if withholding_code == 0:
        rate = 0.23
    elif withholding_code == 1:
        rate = 0.21
    elif withholding_code == 2:
        rate = 0.195
    elif withholding_code == 3:
        rate = 0.185
    else:  # code 4 and above
        rate = 0.18 - ((withholding_code - 4) * 0.005)

    return gross_pay * rate
```

```python
# Test – Person 1: gross pay $750, withholding code 0
gross1 = 750
code1  = 0

soc_sec1  = get_soc_sec_tax(gross1)
medicare1 = get_medicare_tax(gross1)
federal1  = get_federal_tax(gross1, code1)
total1    = soc_sec1 + medicare1 + federal1

print("=== Person 1 ===")
print(f"  Gross Pay:           ${gross1:>8.2f}")
print(f"  Withholding Code:    {code1}")
print(f"  Social Security Tax: ${soc_sec1:>8.2f}  (6.2%)")
print(f"  Medicare Tax:        ${medicare1:>8.2f}  (1.45%)")
print(f"  Federal Tax:         ${federal1:>8.2f}  (23.0%)")
print(f"  Total Tax Withheld:  ${total1:>8.2f}")
```

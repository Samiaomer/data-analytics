```python
import datetime
```

```python
# Today
```

```python
today = datetime.datetime.now()
print(today)
```

```python
print("Today is " + today.strftime("%A, %B %d, %Y"))
```

```python
# My Birthday
```

```python
birthday = datetime.date(1996, 9, 10)  # Change to your actual birthday (year, month, day)
print(birthday)
```

```python
print("My birthday is " + birthday.strftime("%m/%d/%Y"))
```

```python
# Next 90 Days
```

```python
ninety_d = today + datetime.timedelta(days=90)
print(ninety_d)
```

```python
print("90 days from today is " + ninety_d.strftime("%B %d, %Y"))
```

```python
# Time for Dinner
```

```python
dinner_time = datetime.time(7, 30)  # 7:30 PM = 19:30 in 24hr, but we'll format it as 12hr
print(dinner_time)
```

```python
print("Let's meet for dinner at " + dinner_time.strftime("%I:%M %p"))
```

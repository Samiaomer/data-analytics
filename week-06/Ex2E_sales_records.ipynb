This notebook analyzes a small sales dataset using Python functions.
Each function takes the full dataset as its first argument.

```python
sales_data = [
    ('Laptop', 'East', 1200.00),
    ('Phone', 'West', 850.00),
    ('Tablet', 'East', 400.00),
    ('Monitor', 'South', 650.00),
    ('Keyboard', 'West', 75.00),
    ('Laptop', 'South', 1150.00),
    ('Phone', 'East', 900.00),
    ('Tablet', 'West', 425.00),
    ('Monitor', 'East', 680.00),
    ('Keyboard', 'South', 80.00),
    ('Laptop', 'West', 1300.00),
    ('Phone', 'West', 820.00),
]
```

## Lab 1. Total sales by region

```python
def region_total(data, region):
    return sum(amt for _, r, amt in data if r == region)
```

```python
print(region_total(sales_data, 'East'))
print(region_total(sales_data, 'West'))
print(region_total(sales_data, 'South'))
```

## Lab 2. Count of sales by region

```python
def region_count(data, region):
    return sum(1 for _, r, _ in data if r == region)
```

```python
print(region_count(sales_data, 'East'))
print(region_count(sales_data, 'West'))
print(region_count(sales_data, 'South'))
```

## Lab 3. Average sale amount by region

```python
def region_average(data, region):
    count = region_count(data, region)
    if count == 0:
        return 0
    return region_total(data, region) / count
```

```python
print(region_average(sales_data, 'East'))
print(region_average(sales_data, 'West'))
print(region_average(sales_data, 'South'))
print(region_average(sales_data, 'North'))  # Should return 0
```

## Lab 4. Filter by sale amount

```python
def sales_above(data, threshold):
    return [record for record in data if record[2] > threshold]
```

```python
results = sales_above(sales_data, 500)
for record in results:
    print(record)
```

```python
# Bonus
def sales_below(data, threshold):
    return [record for record in data if record[2] < threshold]

results_below = sales_below(sales_data, 500)
for record in results_below:
    print(record)
```

## Lab 5. Top and bottom sale

```python
def top_sale(data):
    return max(data, key=lambda s: s[2])

def bottom_sale(data):
    return min(data, key=lambda s: s[2])

print(top_sale(sales_data))
print(bottom_sale(sales_data))
```

The top sale was a **Laptop** in the **West** region for **$1,300.00**.

The bottom sale was a **Keyboard** in the **West** region for **$75.00**.

## Lab 6. Apply a discount

```python
def apply_discount(data, discount_pct):
    return list(map(lambda s: (s[0], s[1], s[2] * (1 - discount_pct)), data))
```

```python
discounted = apply_discount(sales_data, 0.15)

print('Before discount:')
for record in sales_data[:3]:
    print(record)

print('\nAfter 15% discount:')
for record in discounted[:3]:
    print(record)
```

## Lab 7. Unique regions

```python
def unique_regions(data):
    return sorted(set(r for _, r, _ in data))

print(unique_regions(sales_data))
```

The `unique_regions()` function is useful when writing a longer analysis program because it lets you discover all regions dynamically — without hardcoding any names. For example, you could use this pattern to automatically loop through every region and generate a report for each one, or to populate a dropdown menu in a dashboard. The same pattern applies any time you need a deduplicated list of categories, such as unique product names, salesperson IDs, or time periods.

## Lab 8. Region summary

```python
def region_summary(data, region):
    region_data = [record for record in data if record[1] == region]
    total = region_total(data, region)
    count = region_count(data, region)
    average = region_average(data, region)
    top = top_sale(region_data)

    print(f'=== {region} Region ===')
    print(f'Records: {count}')
    print(f'Total Sales: ${total:,.2f}')
    print(f'Average Sale: ${average:,.2f}')
    print(f'Top Sale: {top[0]} — ${top[2]:,.2f}')

region_summary(sales_data, 'East')
```

## Lab 9. Full summary

```python
for region in unique_regions(sales_data):
    region_summary(sales_data, region)
    print()
```

## Summary

In these labs I practiced writing reusable Python functions that each accept the full dataset as their first argument. Key things I learned:

- **Generator expressions** let you filter and aggregate data in a single concise line inside calls like `sum()`.
- **Building on existing functions** (like calling `region_total()` and `region_count()` inside `region_average()`) keeps code DRY and easier to maintain.
- **`map()` with a lambda** is a clean way to transform every record in a dataset and return a new list without modifying the original.
- **`max()` / `min()` with `key=`** lets you find extreme values in a list of tuples by specifying which element to compare.
- **`set()` for deduplication** combined with `sorted()` is a simple pattern for extracting unique categories from a dataset.
- Combining all these functions in `region_summary()` and `unique_regions()` showed how small, focused functions can be composed into a more complete analysis program.

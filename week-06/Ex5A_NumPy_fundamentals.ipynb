## Exercise Description
In this exercise, we explore the basics of NumPy arrays — including how to create them, check their data types, work with multiple dimensions, and use indexing and transposing.

```python
import numpy as np
```

## Array Basics

```python
# a) Create a standard Python list named 'fruits'
fruits = ['apple', 'banana', 'cherry', 'date']
```

```python
# b) Create a 1D NumPy array from the fruits list and print it
arr_1 = np.array(fruits)
print(arr_1)
```

```python
# c) Check the data type of values contained in arr_1
arr_1.dtype

# The dtype returned is '<U6'.
# The letter 'U' stands for Unicode string — meaning each element is stored as a Unicode character sequence.
# The number '6' is the length (in characters) of the longest string in the array ('banana' and 'cherry' = 6 chars).
# NumPy fixes the string length at the maximum found in the array, so all elements are stored using that many character slots.
```

```python
# d) Create a 2D NumPy array named arr_2 and print it
arr_2 = np.array([[10, 11, 12],
                  [13, 14, 15]])
print(arr_2)
```

```python
# e) Print arr_2 with its dimensions transposed using the .T attribute
print(arr_2.T)
```

```python
# Does .T alter the value stored in arr_2?
print(arr_2)

# No — .T does NOT alter the original array.
# arr_2.T returns a *view* of the array with rows and columns swapped,
# but the underlying data stored in arr_2 remains unchanged (still shape 2x3).
```

```python
# f) Create a 2D NumPy array named arr_3 and print it
arr_3 = np.array([[1, 3, 5],
                  [2, 4, 6]])
print(arr_3)
```

```python
# g) Create arr_4 using slicing notation to extract specific elements from arr_3, then print it
# Element at index position [0, 2] → row 0, column 2 → value 5
# Element at index position [1, 1] → row 1, column 1 → value 4
arr_4 = np.array([arr_3[0, 2], arr_3[1, 1]])
print(arr_4)
```

```python
# h) Create a 3D array of all 1's with two values in each dimension (shape 2x2x2), then print it
arr_5 = np.ones((2, 2, 2))
print(arr_5)
```
